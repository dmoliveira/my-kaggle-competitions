## <span style="color:blue">Airbnb - New User Booking - v3 Training x Validation Analysis</span>

```julia
using Gadfly
using DataFrames
```

```julia
function plot_analysis(dataframe_filename)
    
    df = readtable(dataframe_filename)
    df[:iteration] = vcat(1:size(df, 1));
    
    set_default_plot_size(1000px, 350px)
    
    p1 = plot(df, layer(x=:iteration, y=:rmsetrain, Geom.line), 
              layer(x=:iteration, y=:rmsetest, Geom.line,
              Theme(default_color=colorant"green")),
              Guide.title("RMSE: Train (Blue) x Validation (Green)"))
    
    p2 = plot(df, layer(x=:iteration, y=:mlltrain, Geom.line),
              layer(x=:iteration, y=:mlltest, Geom.line,
              Theme(default_color=colorant"green")),
              Guide.title("MLogLoss: Train (Blue) x Validation (Green)"))
    
    return hstack(p1, p2)    
end
```

```julia
plot_analysis("data/results_training.csv")
```

```julia
plot_analysis("data/results_training_02.csv")
```
